### Задание 1.
Составить в матричном виде двумя способами уравнение поверхности второго порядка $4x^2-24xy+8\sqrt3 xz-7y^2+10\sqrt3 yz+3z^2-4 = 0$. Привести уравнение поверхности второго порядка к каноническому виду с помощью перехода к новому базису. Вывести на экран каноническое уравнение.

```python
from sympy.abc import x, y, z

A = Matrix(
    ((4, -12, 4*3**S.Half), 
    (-12, -7, 5*3**S.Half), 
    ( 4*3**S.Half,  5*3**S.Half, 3)))

X = Matrix((x, y, z))
display(A, Eq(simplify(expand(X.T*A*X))[0] - 4, 0))
P = Matrix([])

for item in A.eigenvects():
    degree = item[1]
    for i in range(degree):
        P = P.row_join(item[2][i].normalized()) 

A2 = P.T*A*P
x1, y1, z1 = symbols('x1 y1 z1')
X_new = Matrix((x1, y1, z1))
q_fRes = simplify(expand(X_new.T*A2*X_new))[0]

display(A2, q_fRes)
eqRes = Eq(q_fRes/4, 1)
display(eqRes)
```

### Задание 2.
Привести уравнение поверхности второго порядка $24 x^{2} - 4 \sqrt{2} x y - 4 \sqrt{2} x z + 18 y^{2} + 12 y z + 18 z^{2} - 4 = 0$ к каноническому виду. Построить график.

```python
from sympy.abc import x, y, z

A = Matrix(((24, -2*2**S.Half, -2*2**S.Half), (-2*2**S.Half, 18, 6), ( -2*2**S.Half,  6, 18)))
X = Matrix((x, y, z))

display(A, Eq(simplify(expand(X.T*A*X))[0] - 4, 0))
P = Matrix([])

for item in A.eigenvects():
    degree = item[1]
    for i in range(degree):
        P = P.row_join(item[2][i].normalized()) 

A2 = P.T*A*P
x1, y1, z1 = symbols('x1 y1 z1')

X_new = Matrix((x1, y1, z1))
q_f2 = simplify(expand(X_new.T*A2*X_new))[0]
display(A2, q_f2)
```

```python
eq2 = Eq(q_f2/4, 1)
display(eq2)

Z = solve(eq2, z1) 
plot3d(*Z, (x1, -1, 1), (y1, -1, 1))
```

### Индивидуальное задание № 139
Привести уравнение поверхности второго порядка к каноническому виду. Построить график.

$u = 13x^2 − 10\sqrt3xy + 23y^2 + 44z^2$


```python
from sympy.abc import x, y, z
A = Matrix(((13, -5*(3**S.Half), 0), (-5*(3**S.Half), 23, 0), (0, 0, 44)))
X = Matrix((x, y, z))

display(A, Eq(simplify(expand(X.T*A*X))[0] -1, 0))
P = Matrix([])

for item in A.eigenvects():
    degree = item[1]
    for i in range(degree):
        P = P.row_join(item[2][i].normalized()) 

# Переход к новому базису
A2 = P.T*A*P
x1, y1, z1 = symbols('x1 y1 z1')
X_new = Matrix((x1, y1, z1))

# Матрицу перехода
q_f2 = simplify(expand(X_new.T*A2*X_new))[0]
display(A2, q_f2)
```

```python
eq2 = Eq(q_f2/2, 1)
display(eq2)
# Находим доп. уравнение
Z = solve(eq2, z1) 
display(*Z)

plot3d(*Z, (x1, -1, 1), (y1, -1, 1))
```
